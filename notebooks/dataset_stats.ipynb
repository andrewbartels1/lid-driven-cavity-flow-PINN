## Report plot generation

After running `pipeline.ipynb` notebook, run this to compute some stats and plots of the dataset(s)

```python
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pnd
from pathlib import Path
```

```python


dataCatalog = pnd.read_csv(Path("../data/ShapeNetCore.v1/datacatalog_parts/datacatalog.csv"))
dataCatalog["simple_label"] = dataCatalog["simple_label"].astype("category")
to_plot = pnd.DataFrame(dataCatalog["simple_label"].value_counts(normalize=False, ascending=False)).round(decimals = 3)
to_plot_norm = pnd.DataFrame(dataCatalog["simple_label"].value_counts(normalize=True, ascending=False)).round(decimals = 3)

to_plot_norm = pnd.DataFrame(dataCatalog["simple_label"].value_counts(normalize=True, ascending=False)).round(decimals = 3)

# get number of variants per simple label
number_var_per_label = dataCatalog.groupby(["simple_label"]).mean().sort_values("numChildren", ascending=False)["numChildren"]



# only plot the first N number, remove number and comment to get all
top_to_plot = 15 # len(to_plot.index)
print(len(to_plot))

# Initialize the matplotlib figure
fig, ax = plt.subplots(figsize=(6, 5))
# Mid point of left and right x-positions
mid = (fig.subplotpars.right + fig.subplotpars.left)/2

ax2 = ax.twiny()
g1 = sns.barplot(x=list(to_plot[0:top_to_plot].simple_label.to_numpy()), y=list(to_plot[0:top_to_plot].index), label="unique label counts", orient = 'h', width=0.8, ax=ax2, alpha=0.5)
g = sns.barplot(x=list(number_var_per_label[0:top_to_plot].to_numpy()), y=list(number_var_per_label[0:top_to_plot].index), color="black", label="Object Variants (\"children\")", orient = 'h', width=0.6, ax=ax)



plt.setp(ax.get_xticklabels(), rotation=80)
ax2.set_xlim([0, 9500])
ax.set_xlim([0, 100])
ax2.bar_label(ax2.containers[0])
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
# ax.set_title("Total number of unique 3D models  {} ".format(len(dataCatalog)))
plt.suptitle("Label & Object Variant Distribution for ShapeNet v.1 \n(first {} categories)".format(top_to_plot),x=mid+0.01)
ax.legend(loc= (.5, .11), frameon = False)
ax2.set_xlabel("Total Number of Models per Category")
ax.set_xlabel("Number of variations per Category")
ax2.legend(loc= (.5, .15), frameon = False)
ax.grid()
plt.tight_layout()
plt.savefig("../images/shapenet_v1_barplot.pdf")
```
